## Preprocess variables

```r
numericVars = which(sapply(data_final2, is.numeric))
numericVars = data_final2[,numericVars] #index vector numeric variables
numericVarNames = names(numericVars) #saving names vector for use later on
numericVarNames
```

```r
#Check skewness of numeric predictors:
skewed_stats = c()
var_names = c()
for (i in 1:ncol(numericVars)){
    skewed_stats = append(skewed_stats,abs(skew(numericVars[,i])))
    var_names = append(var_names, numericVarNames[i])
}

skewed_DF = data.frame(cbind(var_names, skewed_stats))
skewed_DF = skewed_DF[order(skewed_DF$skewed_stats, decreasing = TRUE),]
```

```r
skewed_DF
```

```r
#Set a new dataset before fixing skewness
data_final3 = data_final2
```

```r
# Fix numeric predictors
# Use Box-Cox transformation

skewed_vars = sapply(numericVarNames[-1], function(x) {
    skew(data_final3[[x]], na.rm = TRUE)
})

## Keep only features that exceed a threshold 1 for skewness
skewed_vars = skewed_vars[abs(skewed_vars) > 1]

## Transform skewed features with boxcox transformation
for (x in names(skewed_vars)) {
    bc = BoxCoxTrans(data_final3[[x]], lambda = 0.15)
    data_final3[[x]] = predict(bc, data_final3[[x]])
    # combi[[x]] <- log(combi[[x]] + 1)
}
```

```r
summary(data_final3$SQUARE_FEET)
summary(data_final2$SQUARE_FEET)
```

```r
qqnorm(data_final3$PRICE)
qqline(data_final3$PRICE)
```

```r
## Log transformation of the target variable
data_final3$PRICE = log(data_final3$PRICE + 1)

## Normal distribution after transformation
qplot(PRICE, data = data_final3, bins = 50, main = "Normal distribution after log transformation")
```

```r
qqnorm(data_final3$PRICE)
qqline(data_final3$PRICE)
```

### Convert one hot encoding

```r
factorVars = which(sapply(data_final3, is.factor))
dummies = dummyVars(~., data_final3[names(factorVars)])
data_dmy = predict(dummies, data_final3[names(factorVars)])
data_model = cbind(data_final3, data_dmy)
```

```r
#Drop baseline variables
data_model = within(data_model, rm('BEDS','BATHS','PROPERTY_TYPE','NeighRich'))

dim(data_model)
names(data_model)
```

## Compose Train and Test sets

```r
#Split train set into train and validation set: train 90% and validation 10%
set.seed(123)
train = sample(dim(data_model)[1], dim(data_model)[1]*0.8 )
trainData = data_model[train,]
testData = data_model[-train,]
```

```r
# Split the training data further 
set.seed(123)
train2 = sample(dim(trainData)[1], dim(trainData)[1]*0.9 )
train_set = trainData[train2,]
validation_set = trainData[-train2,]
```

```r
dim(trainData)
dim(testData)
dim(train_set)
dim(validation_set)
```

# Fit Regression Trees

```r
#Using tree package
tree1 = tree(PRICE~., data = train_set)
summary(tree1)
```

```r
plot(tree1)
text(tree1)
```

```r
#Check if pruning the tree will improve performance:
cv.chicago = cv.tree(tree1)
plot(cv.chicago$size, cv.chicago$dev, type = 'b')
```

```r
#Check prediction:
preds1 = predict(tree1, newdata=validation_set)
chicago.test = validation_set$PRICE
plot(preds1, chicago.test)
abline(0,1)
rmse(chicago.test, preds1)
print('MSE')
mean((preds1-chicago.test)^2)
print('Square root of MSE')
sqrt(mean((preds1-chicago.test)^2))
```

```r
#Using rpart package
#This differs from the tree function in S mainly in its handling of surrogate variables. In most
#details it follows Breiman et. al (1984) quite closely. R package tree provides a re-implementation
#of tree.

tree2 = rpart(PRICE ~., data = train_set)
```

```r
rpart.plot(tree2)
```

```r
#Check prediction:
preds2 = predict(tree2, newdata=validation_set)
plot(preds2, chicago.test)
abline(0,1)
print('MSE')
mean((preds2-chicago.test)^2)
print('Square root of MSE')
sqrt(mean((preds2-chicago.test)^2))
#cat('This model leads to test predictions that are within around $',sqrt(mean((preds2-chicago.test)^2)), 'of the true median home value for the area')
```

# Try Boosting

```r
set.seed(1)
boost.chicago1 = gbm(PRICE ~., data = train_set, distribution ="gaussian", n.trees = 1000, interaction.depth = 6 )
```

```r
#Produce relative influence plot and the relative influence statistics
summary(boost.chicago1)
```

```r
#Produce partial dependence plots
#For SQUARE_FEET and BATHS
par(mfrow = c(1,2))
plot(boost.chicago1, i = 'SQUARE_FEET')
plot(boost.chicago1, i = 'BATHS')
```

```r
#Predict validation set
preds.boost1 = predict(boost.chicago1, newdata = validation_set, n.trees = 1000)
print('MSE')
mean((preds.boost1 - chicago.test)^2)
print('Square root of MSE')
sqrt(mean((preds.boost1 - chicago.test)^2))

rmse(chicago.test, preds.boost1)
```

```r
#Adjust shrinkage parameter
boost.chicago2 = gbm(PRICE ~., data = train_set, distribution ="gaussian", n.trees = 1000, interaction.depth = 6,
                    shrinkage = 0.2, verbose = F)
```

```r
preds.boost2 = predict(boost.chicago2, newdata = validation_set, n.trees = 1000)
print('MSE')
mean((preds.boost2 - chicago.test)^2)
print('Square root of MSE')
sqrt(mean((preds.boost2 - chicago.test)^2))
```

# Try XGBOOST

```r
set.seed(123)
xgb.chicago1 = xgboost(data = as.matrix(train_set[, -1]), nfold = 5, label = as.matrix(train_set$PRICE), 
    nrounds = 2200, verbose = FALSE, objective = "reg:linear", eval_metric = "rmse", 
    nthread = 8, eta = 0.01, gamma = 0.0468, max_depth = 6, min_child_weight = 1.7817, 
    subsample = 0.5213, colsample_bytree = 0.4603)
```

```r
preds.xgb1 = predict(xgb.chicago1, newdata = as.matrix(validation_set[, -1]))
rmse(chicago.test, preds.xgb1)
print('Square root of MSE')
sqrt(mean((preds.xgb1 - validation_set$PRICE)^2))
```

```r
#Tune in parameters
xgb_grid = expand.grid(
nrounds = 1000,
eta = c(0.1, 0.05, 0.01),
max_depth = c(2, 3, 4, 5, 6),
gamma = 0,
colsample_bytree=1,
min_child_weight=c(1, 2, 3, 4 ,5),
subsample=1
)

my_control = trainControl(method="cv", number=5)
```

```r
xgb_caret =  train(x=train_set[,-1], y=train_set$PRICE, method='xgbTree', trControl= my_control, tuneGrid=xgb_grid) 
xgb_caret$bestTune
```
